### <span style="color:lightblue"> Categorize Geography </span>
Categorize using One Hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder()
geography_encoded = ohe.fit_transform(dataset[['Geography']]).toarray()
print(ohe.get_feature_names_out(['Geography']))
geography_encoded
```

```python
geo_ohe_df = pd.DataFrame(geography_encoded, columns=ohe.get_feature_names_out(['Geography']))
geo_ohe_df
```

```python
dataset = pd.concat((dataset.drop("Geography", axis=1), geo_ohe_df), axis=1)
dataset.head()
```

### <span style="color:yellow"> Devide Data into Features </span>

<span style="color:lightblue;">1. Independent Feature (X)</span>

- These are <span style="color:#117A65;">input variables</span> used to predict or explain another variable.
- Also known as <span style="color:#AF601A;">predictors</span>, <span style="color:#AF601A;">features</span>, or <span style="color:#AF601A;">independent variables</span>.
- They are not influenced by other variables in the dataset.
- In a churn dataset, examples include:
  - `Age`, `CreditScore`, `Balance`, `Gender`, `Tenure`, etc.

<br>

<span style="color:green;">2. Dependent Feature (y)</span>

- This is the <span style="color:#117A65;">target variable</span> you want to predict or explain.
- Also known as <span style="color:#AF601A;">label</span>, <span style="color:#AF601A;">output</span>, or <span style="color:#AF601A;">dependent variable</span>.
- Its value is affected by the independent features.
- In a churn dataset, a typical dependent feature is:
  - `Exited` → whether the customer has left the bank (1 = yes, 0 = no).


```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

ind_features = dataset.drop(['Exited'], axis=1)
d_features = dataset['Exited']

# split data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(ind_features, d_features, test_size=0.33, random_state=42)

# Scale the features
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

x_train
x_test
```

`import pickle`

Purpose:
To store the scaler after it has been fitted on the training data, so you can reuse it later without refitting.

This ensures consistent preprocessing during inference or deployment — especially important in machine learning pipelines.

It's useful when deploying models, sharing projects, or reloading the same environment across different sessions.

```python
import pickle

with open('dataScaler.pkl', 'wb') as file:
    pickle.dump(scaler, file)
```

```python
with open('gender_encoder.pkl', 'wb') as file:
    pickle.dump(le, file)
        
with open('geography_encoder.pkl', 'wb') as file:
    pickle.dump(ohe, file)
```

### <span style="color:red"> ANN Implementation </span>

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Build ANN model
model = Sequential([
    Dense(128, activation='relu', input_shape=(x_train.shape[1],)),
    Dense(64, activation='relu'),
    Dense(1, activation='sigmoid')
])

model.summary()
```

`tf.keras.optimmizers.Adam()`
chose Adam because it provides the best convergence speed, stability, and adaptivity for most cases, including binary classification. That is why Adam is the default optimizer in many Keras models.

```python
# Train Model for Binary Classification
optimize = tf.keras.optimizers.Adam(learning_rate=0.001)
loss = tf.keras.losses.BinaryCrossentropy()

# Compile the model
model.compile(optimizer=optimize, loss=loss, metrics=['accuracy'])
```

Key Functions of Optimizer

Weight Adjustment: Updates the model's weights (parameters) iteratively based on the loss function's gradient with respect to the weights.

Minimizing Loss: The optimizer reduces the error (or loss) between the predicted and actual outputs.

Convergence: Ensures the training process converges the best possible weights that make the model as accurate as possible

### <span style="color:lightblue"> SetUp tensorBoard for Monitoring </span>

```python
import tensorflow as tf
print(tf.executing_eagerly())  # Harus True
```

```python
# Setting up TensorBoard

from tensorflow.keras.callbacks import TensorBoard, EarlyStopping

log_dir = "log/modelProcessing"
tensorflow_calback = TensorBoard(log_dir=log_dir, histogram_freq=1)

# Setting up Early Stopping
early_stopping_callback = EarlyStopping(monitor='val_loss', patience=20, restore_best_weights=True)

# Train the model
model_training = model.fit(
    x_train, y_train, validation_data=(x_test, y_test),epochs=100,
    callbacks=[tensorflow_calback, early_stopping_callback]
)
```
